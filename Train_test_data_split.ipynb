```python
%matplotlib inline
# Imports Packages
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import seaborn as sns
import sklearn
import numpy as np
from sklearn import metrics
```

```python
# Import data
NBE_df = pd.read_excel('ML_Model_Data_2021.xls', sheet_name='Base')

#Filter Element Data
NBE_element_df = NBE_df[NBE_df['Steel_Element'] == 107]

# print(NBE_element_df)
NBE_element_df.head()
```

```python
# Data Split
# Identify Feature Importance for Decision Tree Regression

#NBI Dataset
X = NBE_df.loc[:,['Bridge_Number', 'Steel_Element', 'Coating_type', 'MAINTENANCE_021', 'Bridge_Age', 'ADT_029', 'DEGREES_SKEW_034',
       'SERVICE_ON_042A', 'SERVICE_UND_042B', 'STRUCTURE_KIND_043A',
       'STRUCTURE_TYPE_043B', 'APPR_KIND_044A', 'APPR_TYPE_044B',
       'DECK_COND_058', 'SUPERSTRUCTURE_COND_059', 'SUBSTRUCTURE_COND_060',
       'OPERATING_RATING_064', 'INVENTORY_RATING_066', 'STRUCTURAL_EVAL_067',
       'APPR_ROAD_EVAL_072', 'DECK_STRUCTURE_TYPE_107', 'SURFACE_TYPE_108A',
       'MEMBRANE_TYPE_108B', 'DECK_PROTECTION_108C', 'PERCENT_ADT_TRUCK_109',
       'BRIDGE_CONDITION']]
Y = NBE_df.loc[:,['Bridge_Number', 'Steel_Element', 'Coating_type', 'CS1', 'CS2','CS3','CS4']]

# split into train test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```
